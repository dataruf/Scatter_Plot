```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df = pd.read_csv(r"C:\Users\User\Desktop\GitHub\DataVisualization\Scatter Plot\data\scatter_data.csv")
```

```python
df.head()
```

```python
# # Ərazi və Qiymət arasısndakı əlaqənin mövcüdluğunu, gücü və istiqamətini öyrənmək üçün 
# scatter plot vizualından istifadə edəcəyik
```

```python
sns.set_style("darkgrid")
```

```python
plt.figure(figsize = (13,7))
scatter = plt.scatter(df['Area (ft.)'],
                      df['Price'],
                      alpha = 0.3,
                      c = df['Building Type'],
                      cmap = 'viridis')
plt.legend(*scatter.legend_elements(),
          loc = 'upper left',
          title = 'Building Type')
plt.title("Connection between Price and Area", fontsize = 14, weight = "bold")
plt.xlabel("Area size (ft.)")
plt.ylabel("Price")
sns.despine()
plt.savefig("Scatter_Plot_Visual.png")
plt.show()
```
